```python
import os
import json
from urllib.request import urlretrieve

url = "http://faculty.cs.niu.edu/~dakoop/cs503-2024fa/a3/food-data-sample.json"
local_fname = 'food-data-sample.json'
if not os.path.exists(local_fname):
    urlretrieve(url, local_fname)
data = json.load(open(local_fname))
```

# 1. Serving Size Units (5 pts)

```python
serving_size_list = []

for item in data:
    # if unit doesnt exist in list, add it
    if item['serving_size_unit'] not in serving_size_list:
        serving_size_list.append(item['serving_size_unit'])

print(serving_size_list)
```

# 2. Largest Serving Size (5 pts)

```python
largest = []

# Get all items that have ml as the serving size unit
for item in data:
    if item['serving_size_unit'] == 'ml':
        largest.append((item['brand_owner'], item['description'], item['serving_size']))

# Use sorted method, using a lambda to specify the serving size unit
largest = sorted(largest, reverse=True, key=lambda brand: brand[2])

# Print out the brand, desc, and serving size
for brand in largest:
    print(f"{brand[0]} {brand[1]}")

# I hope this is right, the question said nothing about a limit
```

# 3. Category Counts (10 pts)

```python
categories_count = {}

for item in data:
    category = item["branded_food_category"]
    # if the category is in the list, add one. if not, new entry
    if category in categories_count:
        categories_count[category] += 1
    else:
        categories_count[category] = 1

# Find the max categories using the max function
max_category = max(categories_count, key=categories_count.get)

print(max_category)
```

# 4. Add Unsaturated Fat (15 pts)

```python
for item in data:
    fat = {'Saturated Fat': None, 'Total Fat': None}
    unsat_fat = 0

    # Loop through nutrition list
    for i in item['nutrition']:
        if i['name'] in fat:
            fat[i['name']] = i['amount']

    # If either is none, the satur fat is none
    if fat['Saturated Fat'] == None or fat['Total Fat'] == None:
        unsat_fat = None
    else:
        unsat_fat = fat['Total Fat'] - fat['Saturated Fat']
        # Negative values don't make sense
        if unsat_fat < 0:
            unsat_fat = 0
    # Append new value to nutrition list
    item['nutrition'].append({'name': 'Unsaturated Fat', 'amount': unsat_fat, 'unit_name': 'G'})

data[0]['nutrition']
```

# 5. Filter by Fiber Range (15 pts)

```python
def filter_by_fiber(min_fiber, max_fiber):
    filtered_items = []
    
    for item in data:
        for i in item['nutrition']:
            if i['name'] == 'Fiber':
                if  min_fiber <= i['amount'] <= max_fiber:
                    filtered_items.append(item['description'])

    print(filtered_items)

filter_by_fiber(7.3, 7.35)
```

# 6. Filter by Ingredients (Extra Credit)

```python
def filter_by_ingredients(x):
    filtered_items = []
    ingredient = x.upper()
    
    for item in data:
        current_ingredient_list = item.get('ingredients')

        # If exists and ingredient is in list, add the item to the return list
        if current_ingredient_list and ingredient in current_ingredient_list:
            filtered_items.append(item)

    return filtered_items

print(len(filter_by_ingredients('apple')))
[d['description'] for d in filter_by_ingredients('saffron')]
```
